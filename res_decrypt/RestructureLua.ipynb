```python
for root, dirs, files in os.walk(decDir):
    #print(root, files, dirs)
    for f in files:
        if not f.endswith('.luac_dec'):
            continue
        path = f'{root}/{f}'
        print(path)
        
        firstLine = open(path).readline()
        PREFIX = ' -- source: '
        if firstLine.startswith(PREFIX):
            luapath = firstLine[len(PREFIX):].strip() + '.lua'
            targetpath = f'{outDir}/{luapath}'
        else:
            luafileid = os.path.basename(path)[:-len('.luac_dec')]
            luapackageid = os.path.basename(os.path.dirname(path))
            if luafileid in mappingDir:
                targetpath = mappingDir[luafileid]
            elif luapackageid in mappingDir:
                targetpath = f'{mappingDir[luapackageid]}/{luafileid}.luac_dec'
            else:
                oripath = os.path.relpath(path, decDir)
                targetpath = f'{outDir}/{oripath}'
        os.makedirs(os.path.dirname(targetpath), exist_ok=True)
        shutil.copy(path, targetpath)
```
